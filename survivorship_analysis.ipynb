```python
import requests

df = pd.read_csv('sorted_apps.csv')

google_df = df[df['store'] == 'Google Play'].copy()
chrome_df = df[df['store'] == 'Chrome Web Store'].copy()
apple_df = df[df['store'] == 'Apple App Store'].copy()


apple_df['alive'] = False  # create column
dead_urls = []  # List to store dead URLs


for i, (idx, row) in enumerate(apple_df.iterrows(), 1):
    url = row['url']
    name = row['title']

    try:
        response = requests.get(url, timeout=10)
        alive = response.status_code == 200
    except Exception as e:
        alive = False

    apple_df.at[idx, 'alive'] = alive

    if not alive:
        dead_urls.append(url)  # Add dead URL to the list
        print(f"[{i}] DEAD - {name} (Apple App Store)")
        
print("✔️ Apple survivorship done and saved.")


# Filter apps 
dead_apps = apple_df[apple_df['alive'] == False]
alive_apps = apple_df[apple_df['alive'] == True]

print(f"💀 Saved {len(dead_apps)} dead apps from apple app store, total {len(apple_df)} apps")

```

```python
google_df['alive'] = False

for i, (idx, row) in enumerate(google_df.iterrows(), 1):
    url = row['url']
    name = row['title']

    try:
        response = requests.get(url, timeout=10)
        alive = response.status_code == 200
    except Exception as e:
        alive = False

    google_df.at[idx, 'alive'] = alive

    if not alive:
        dead_urls.append(url)  # Add dead URL to the list
        print(f"[{i}] DEAD - {name} (goole play)")
        
print("✔️ google play survivorship done and saved.")


# Filter apps 
google_dead_apps = google_df[google_df['alive'] == False]
google_alive_apps = google_df[google_df['alive'] == True]

print(f"💀 Saved {len(google_dead_apps)} dead apps from apple app store, total {len(google_df)} apps")

```

```python
# Combine dead apps from Apple and Google Play
dead_apps = pd.concat([dead_apps, google_dead_apps], ignore_index=True)

alive_apps = pd.concat([alive_apps, google_alive_apps], ignore_index=True)

print(f"🧟‍♂️ Total dead apps (Apple + Google): {len(dead_apps)}")
print(f"🟢 Total alive apps (Apple + Google): {len(alive_apps)}")
```

```python
chrome_df['alive'] = False

for i, (idx, row) in enumerate(chrome_df.iterrows(), 1):
    url = row['url']
    name = row['title']

    print(f"[{i}/{len(chrome_df)}] Checking: {name} - {url}")
    
    try:
        response = requests.get(url, timeout=10)
        alive = response.status_code == 200
    except Exception as e:
        alive = False
        print(f"❌ Error checking {name}: {e}")

    chrome_df.at[idx, 'alive'] = alive

    if not alive:
        dead_urls.append(url)  # Add dead URL to the list
        print(f"[{i}] DEAD - {name} (chrome extension)")
        
print("✔️ chorme extension survivorship done and saved.")


# Filter apps 
chrome_dead_apps = chrome_df[chrome_df['alive'] == False]
chrome_alive_apps = chrome_df[chrome_df['alive'] == True]

print(f"💀 Saved {len(chrome_dead_apps)} dead apps from chrome extension, total {len(chrome_df)} apps")

```

```python
dead_apps = pd.concat([dead_apps, chrome_dead_apps], ignore_index=True)

alive_apps = pd.concat([alive_apps, chrome_alive_apps], ignore_index=True)

print(f" Total dead apps : {len(dead_apps)}")
print(f" Total alive apps : {len(alive_apps)}")
```
