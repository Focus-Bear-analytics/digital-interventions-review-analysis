```python
from google_play_scraper import search
import time

keywords = [
    "distraction",
    "addiction",
    "motivation",
    "self-control",
    "smartphone distraction",  
    "smartphone addiction",
    "smartphone motivation",
    "smartphone self-control",
    "laptop distraction",
    "laptop addiction",
    "laptop motivation",
    "laptop self-control",
    "internet distraction",
    "internet addiction",
    "internet motivation",
    "internet self-control"
    
    
]

# Create a dictionary to store results for each keyword
all_results = {}

# Loop through each keyword in the list and search
for keyword in keywords:
    print(f"Searching for: {keyword}")
    try:
        # Perform search
        results = search(keyword, lang="en", country="au")
        
        # Store only the specific features for each app
        filtered_results = []
        for app in results:
            app_info = {
                'title': app['title'],
                "appId": app["appId"],
                'genre': app['genre'],
                'installs': app['installs'],
                'price': app['price'],
                'free': app['free'],
                'description': app['description']
            }
            filtered_results.append(app_info)
            
            # Print the specific features
           # print(f"App: {app['title']} | genre: {app['genre']} | installs: {app['installs']} | price: {app['price']} | free: {app['free']} | description: {app['description']}" )

                # Store the filtered results for this keyword
        all_results[keyword] = filtered_results
        
        # Wait a second between searches to avoid hitting the server too fast
        time.sleep(1)
        
    except Exception as e:
        print(f"Error with keyword '{keyword}': {e}")

# issues 
# only for 30 each maximum. 
```

```python
import csv

# Define the CSV file path
csv_file_path = "app_search_results.csv"

# Define the CSV headers
headers = ["keyword", "title", "appId", "genre", "installs", "price", "free", "description"]

# Open the CSV file for writing
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    
    # Write the header row
    writer.writeheader()
    
    # Iterate through each keyword and its results
    for keyword, apps in all_results.items():
        for app in apps:
            # Create a row with both keyword and app data
            row = {
                "keyword": keyword,
                "title": app["title"],
                "appId": app["appId"],
                "genre": app["genre"],
                "installs": app["installs"],
                "price": app["price"],
                "free": app["free"],
                "description": app["description"]
            }
            
            # Write the row to the CSV file
            writer.writerow(row)

print(f"CSV file has been created at {csv_file_path}")
```

```python
from google_play_scraper import reviews, Sort
import csv
import time

input_file = "app_search_results_cleaned.csv"
output_file = "adhd_apps_with_reviews.csv"
adhd_keyword = "adhd"

# Open input and output CSV files
with open(input_file, 'r', newline='', encoding='utf-8') as infile, \
     open(output_file, 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['reviews']  # Add a column for the reviews
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for row in reader:
        app_id = row['appId']
        print(f"ğŸ“ Fetching reviews for: {row['title']} ({app_id})")
        matched_reviews = []

        try:
            # Fetch reviews with default sorting (newest first)
            review_data, _ = reviews(
                app_id,
                lang='en',
                country='au',
                sort=Sort.NEWEST,  # Sorting by newest
            )

            # Filter the reviews to find ones with the "adhd" keyword
            for r in review_data:
                if adhd_keyword in r['content'].lower():
                    matched_reviews.append(r['content'])  # Add the review content directly

        except Exception as e:
            print(f"âš ï¸ Error fetching reviews for {app_id}: {e}")

        # Only write to CSV if there are matched reviews
        if matched_reviews:
            row['reviews'] = "\n---\n".join(matched_reviews)  # Join matched reviews with a separator
            writer.writerow(row)  # Write the row only if there are reviews
            print(f"âœ”ï¸ Reviews for {app_id} saved.")

        time.sleep(1)

print(f"\nâœ… Reviews with ADHD keyword saved to: {output_file}")
```
